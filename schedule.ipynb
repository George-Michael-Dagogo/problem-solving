```python
import schedule
import pandas as pd
import time
from datetime import datetime
import random
# datetime object containing current date and time
from bs4 import BeautifulSoup
import requests
url = "https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBM-DA0321EN-SkillsNetwork/labs/datasets/Programming_Languages.html"
data =  requests.get(url).text
soup = BeautifulSoup(data, 'html5lib')
```

```python
def omni_time():
    now = datetime.now()
    # dd/mm/YY H:M:S
    dt_string = now.strftime("%d/%m/%Y %H:%M:%S")
    return dt_string

def webme():
    my_dict = {"Language":[],"Salary":[]};

    for row in soup.find_all('tr'):
        dat = row.find_all('td')
        lang= dat[1].getText()
        pay = dat[3].getText()
        my_dict["Language"].append(lang)
        my_dict["Salary"].append(pay)
    
    df = pd.DataFrame(my_dict)
    yeah = df[1:]
    from sqlalchemy import create_engine
    engine = create_engine('sqlite:///save_pandas.db', echo=True)
    sqlite_connection = engine.connect()
    
    sqlite_table = "forex_predictor"
    yeah.to_sql(sqlite_table, sqlite_connection, if_exists='replace')
    
    sqlite_connection.close()
    
    
    




    
```

```python
schedule.every(30).seconds.do(webme)
#to make the schedule run every two seconds
#schedule.every().day.at('12:00').do(webme)
#runs a code at a particular time of day
while True:
    schedule.run_pending()
    time.sleep(1)
```
